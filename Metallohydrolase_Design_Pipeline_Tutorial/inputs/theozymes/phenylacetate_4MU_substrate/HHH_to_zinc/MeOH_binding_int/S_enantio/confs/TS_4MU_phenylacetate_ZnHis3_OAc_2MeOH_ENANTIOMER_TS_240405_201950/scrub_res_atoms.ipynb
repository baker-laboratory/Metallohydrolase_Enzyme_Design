```python
xyzf = open("saved_xyz.xyz", "r").readlines()
```

```python
xyzs = []
for l in xyzf:
    if l == "86\n":
        xyzs.append([])
    xyzs[-1].append(l)
```

```python
len(xyzs)
```

```python
keep_atom_ranges = [[1, 18], [66, 86]]
xyz_cropped = []
for xyz in xyzs:
    xyz_cropped.append([xyz[1]])
    for rng in keep_atom_ranges:
        for n in range(rng[0], rng[1]+1):
            xyz_cropped[-1].append(xyz[n+1])
```

```python
xyz_cropped[0]
```

```python
with open("saved_xyz_cropped.xyz", "w") as file:
    for xyz in xyz_cropped:
        file.write(f"{len(xyz)-1}\n")
        for l in xyz:
            file.write(l)
```
